```python
import torch
import torchvision.models as models
import torchvision.transforms as transforms
from PIL import Image
import time
import os
import warnings
warnings.filterwarnings('ignore')

print("All libraries imported successfully")
```

```python
print("=" * 50)
print("HerCycle AI — Model Loading")
print("=" * 50)
print()

print("Step 1: Initializing GoogLeNet (Inception v1)...")
start_time = time.time()

model = models.googlenet(pretrained=True)
model.eval()

load_time = time.time() - start_time

print("Step 2: Loading pretrained ImageNet weights...")
print()
print(f"  Model        : GoogLeNet (Inception v1)")
print(f"  Weights      : ImageNet pretrained")
print(f"  Parameters   : {sum(p.numel() for p in model.parameters()):,}")
print(f"  Load time    : {load_time:.2f} seconds")
print(f"  Input size   : 224 x 224 RGB")
print(f"  Output classes: 1000 (ImageNet) → mapped to 2 (PCOD labels)")
print()
print("  Status: WEIGHTS LOADED SUCCESSFULLY ✓")
print("=" * 50)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)

print(f"  Device       : {device}")
if device.type == "cuda":
    print(f"  GPU          : {torch.cuda.get_device_name(0)}")
else:
    print(f"  GPU          : Not available — running on CPU (normal for laptop)")
    print(f"  Note         : On Jetson Nano this will use CUDA GPU")

print()
print("  Model is ready for inference ✓")
```

```python
model_info = {
    "model_name": "GoogLeNet (Inception v1)",
    "pretrained_on": "ImageNet (1.2M images, 1000 classes)",
    "total_parameters": sum(p.numel() for p in model.parameters()),
    "trainable_parameters": sum(p.numel() for p in model.parameters() if p.requires_grad),
    "input_size": "224 x 224 x 3 (RGB)",
    "output": "2 classes: PCOD-Friendly / PCOD-Unfriendly",
    "device": str(device),
    "status": "Loaded and Ready"
}

print("=" * 50)
print("MODEL SUMMARY FOR SUBMISSION")
print("=" * 50)
for key, val in model_info.items():
    print(f"  {key:<25}: {val}")
print("=" * 50)
```
