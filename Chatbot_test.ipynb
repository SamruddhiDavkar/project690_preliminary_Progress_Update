```python
import ollama
import time

print("Ollama library ready ✓")
```

```python
questions = [
    "What foods should someone with PCOD eat?",
    "Is white rice bad for PCOD?",
    "What supplements help with PCOD?"
]

system_prompt = """You are HerCycle AI, a PCOD wellness assistant. 
Only answer questions about PCOD diet, nutrition, and lifestyle.
If asked about medication or diagnosis, redirect to a doctor.
Keep answers concise and practical."""

print("=" * 55)
print("HerCycle AI — Chatbot Speed Test")
print("=" * 55)

for q in questions:
    start = time.time()
    
    response = ollama.chat(
        model='llama3.2:3b-instruct-q4_K_M',
        messages=[
            {'role': 'system', 'content': system_prompt},
            {'role': 'user',   'content': q}
        ]
    )
    
    elapsed = time.time() - start
    answer = response['message']['content']
    tokens = response['eval_count']
    speed  = tokens / elapsed

    print(f"\nQ: {q}")
    print(f"A: {answer[:200]}...")
    print(f"\n  ⏱ Time     : {elapsed:.1f}s")
    print(f"  🔢 Tokens   : {tokens}")
    print(f"  ⚡ Speed    : {speed:.1f} tokens/sec")
    print("-" * 55)
```
